```python
x=np.linspace(0.0,100.0,1001)
y=x*x
plt.plot(x,y,'r-')
```

```python
x=np.linspace(0.0,100.0,1001)
y1=x*x
y2=2*(x*x)
y3=1-(x*x)
plt.plot(x,y1,'r-')
plt.plot(x,y2,'b-')
plt.plot(x,y3,'g-')
```
